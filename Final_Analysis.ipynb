```python
def extract_state(location):

    if len(location) == 3 :
        return location[-2]
    else :
        return location [-1]
```

```python
df_au=df[df["Country"]=="AU"]
```

```python
df_au["State"] = df_au["Location"].str.split("^").apply(extract_state)
```

```python
def clean_brand(brand_str):
    return str(brand_str).upper()

df_au["Brand"] = df_au["Brand"].apply(clean_brand)
    
```

```python
df_au['Brand'].value_counts()[:6]
```

```python
def clean_price(price_str):
    try : 
         price = float(price_str)
    except:
        try : 
            price = float(''.join(ele for ele in price_str if ele.isdigit() or ele == '.'))
        except:
            price=0
    return price
#df_au.loc[df_au["Price"]].apply(clean_price)
df_au["Price"] = df_au["Price"].apply(clean_price)
```

```python
df_au['Capacity'].value_counts()
```

```python
# Get rid of L in Capacity
def remove_Litr(capacity_str):
    if "L" in str(capacity_str):
        return str(capacity_str)[:capacity_str.find("L")].strip()
    else:
        return capacity_str

df_au['Capacity']=df['Capacity'].apply(remove_Litr)
```

```python
df_au['Capacity'].value_counts()
```

```python
def get_capacity_group(cap):
    if cap==None :
        return "Nan"
    try :
        cap = int(cap)
        res = ">600"
        if cap < 200 : 
            res = "<200"
        elif cap >= 200 and cap < 250 :
            res = "200-250"
        elif cap >= 250 and cap < 300 :
            res = "250-300"
        elif cap >= 300 and cap < 350 :
            res = "300-350"
        elif cap >= 350 and cap < 400 :
            res = "350-400"
        elif cap >= 400 and cap < 450 :
            res = "400-450"
        elif cap >= 450 and cap < 500 :
            res = "450-500"
        elif cap >= 500 and cap < 550 :
            res = "500-550"
        elif cap >= 550 and cap < 600 :
            res = "550-600"
       
            
    except:
        res = "Nan"
    return res    
        
df_au["Capacity_Type"] = df_au["Capacity"].apply(get_capacity_group)
```

```python
df_au["Capacity_Type"].value_counts()
```

```python
df_result = df_au[["Price","State","Brand","Capacity_Type"]].groupby(["State","Brand","Capacity_Type"]).describe()
df_result
```
